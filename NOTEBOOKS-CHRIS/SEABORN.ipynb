```python
# This function shall help us plot some offset sine waves:
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)

with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

```python
names = ['group_a', 'group_b', 'group_c', 'grupo_d']
values = [1, 10, 50, 80]
#plt.subplots entrega una tupla (la figura, los ejes)
figura, axes = plt.subplots(1,2, figsize=(10,5), dpi=85)

#axes es una fila con dos columnas (1,2) de acuerdo a eso podemos incluir la grafica

#Colocando titulo general
figura.suptitle("Ejemplo de Graficas de Barras")

#Dibujamos la grafica 
sns.barplot(x=values, y=names, ax=axes[0])
axes[0].set_xlabel("Cantidad")
axes[0].set_ylabel("Grupos")
axes[0].set_title("Cantidad por Grupo") 

sns.barplot(x=names, y=values, ax=axes[1])
axes[1].set_xlabel("Cantidad")
axes[1].set_ylabel("Grupos")
axes[1].set_title("Cantidad por Grupo") 

sns.set_style("dark") #Con esto le damos tema al fondo de las graficas

figura.tight_layout() #Esta funcion nos permite acomodar el Layaout y evitar traslapes
```

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 
  
# cargando un dataset
data = sns.load_dataset("iris") 
  
# dibujando el lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# seteando el titulo con matplotlib
plt.title('Title using Matplotlib Function') #Colocando titulo
plt.xlim(5) #Hace que el limite de eje x empiece en 5
#Por default es 0
#plt.ylim tbm existe

# Cambiando el tema a white, Existe los temas: dargrid, whitegrid, dark, white, ticks
sns.set_style("white")

#Cambiando el tamaño de la grafica
plt.figure(figsize=(5,5))

# Quitando los marcos del grafico con la funcion despine
sns.despine(left=True) #Tiene mas opciones para el grafico o subgraficos

plt.show() #Esto tambien ayuda o no tener textos como ouput
```
