```python
import pandas as pd

filepath = "C:/Users/Usuario/Desktop/CVC Temporal/Zona Norte - Setiembre 2022 - Olmos.xlsx"

df_cvc = pd.read_excel(io=filepath, sheet_name="BD", index_col=0, header=1) #Leemos el dataset

df_cvc.set_index(pd.Index(list(range(len(df_cvc)))), inplace=True) #Agregamos un indice a todo el datfram

df_cvc.head()
```

```python
df_cvc.tail()
```

```python
columnas = df_cvc.columns.values.tolist() #Obtenemos todas las columnas
columnas
```

```python
df_cvc['Año'] = df_cvc["Mes"].apply(lambda x: str(x)[:4]) #Creando una columna usando los 4 primeros digitos de la columna Mes

df_cvc["Mes_"] = df_cvc["Mes"].apply(lambda x: str(x)[4:]) #Creamos una columna usando los digitos restantes de la columna Mes

columnas = ["Año", "Mes_"] + columnas[1:] #Hacemos que las columnas nuevas se agreguen a la lista de columnas y quitanto Mes
columnas
```

```python
for ind,col in enumerate(columnas):
    print(ind,col)
```

```python
col_select = columnas[0:8] + columnas[14:19] + columnas[20:22] + columnas[74:75]
col_select
```

```python
df_cvc1 = df_cvc[col_select]
```

```python
df_cvc1.head()
```

```python
for col in df_cvc1.columns.values.tolist(): #Averiguando que columnas tienes valores Nulos
    if df_cvc1[col].isnull().sum()>0:
        print(col)
```

```python
df_cvc1.isnull().sum() #Se muestra que en la columna Ruc hay 153 nulos (NAN, NAT, NULL)
```

```python
df_cvc1.fillna(value={"Ruc":"00000000000"}, inplace=True) #Rellenando con ceros
```

```python
df_cvc1.isnull().sum() #Se observa que ya no hay datos faltantes
```

```python
#Filtrando solo aquellos que son Clientes Libres 
df_cvc1_libres = df_cvc1[df_cvc1["Tipo de cliente"]=="Libre"]
df_cvc1_libres
```

```python
df_cvc1_Reg = df_cvc1[df_cvc1["Tipo de cliente"]=="Regulado"]
df_cvc1_Reg
```

```python
df_agrupado_libres = df_cvc1_libres.groupby(by=["Año", "Mes_"])
```

```python
df_agrupado_libres.first()
```

```python
df_agrupado_libres.get_group(("2018","02")) #Mostrando uno de los grupos
```

```python
df_agrupado_reg = df_cvc1_Reg.groupby(by=["Año", "Mes_"])
df_agrupado_reg.first()
```

```python
import numpy as np
#s = lambda x: x.sum()/1000
df_kwh_total_libres = df_agrupado_libres.agg({"Energía Activa Total": lambda x: round(x.sum(),2), 
                                              "Total": np.sum,
                                             "Energía Hora Punta": lambda x: round(x.sum(),2),
                                             "Energía Fuera de Punta": lambda x: round(x.sum(),2)
                                             }) #La columnas de energia y total facturado
#Se lo asignaremos a una variable
```

```python
df_kwh_total_libres #Esto es un tipo groupby el cual es muy util
```

```python
df_kwh_total_libres.rename(columns={"Energía Activa Total":"EAT", 
                                    "Total":"FacTotal",
                                    "Energía Hora Punta":"EAHP",
                                    "Energía Fuera de Punta":"EAHFP"}, inplace=True)
df_kwh_total_libres
```

```python
df_kwh_total_reg = df_agrupado_reg.agg({"Energía Activa Total": lambda x: round(x.sum(),2), 
                                        "Total": np.sum,
                                        "Energía Hora Punta": lambda x: round(x.sum(),2),
                                        "Energía Fuera de Punta": lambda x: round(x.sum(),2)})
df_kwh_total_reg
```

```python
df_kwh_total_reg.rename(columns={"Energía Activa Total":"EAT", 
                                    "Total":"FacTotal",
                                    "Energía Hora Punta":"EAHP",
                                    "Energía Fuera de Punta":"EAHFP"}, inplace=True)
df_kwh_total_reg
```
