# Agregación de datos por categoría

```python
import numpy as np
import pandas as pd
```

```python
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
n = 500 #Numero de filas
gender_data = []
income_data = []

for i in range(0,n):
    gender_data.append(np.random.choice(gender)) #np.random.choice() es para escoger entre los valores de una lista
    income_data.append(np.random.choice(income))
```

```python
#Z -> N(0,1)
#N(m,s) -> m + s*Z
height = 160 + 30*np.random.randn(n)  #np.random.randn()
weight = 65 + 25*np.random.randn(n)
age = 30 + 12*np.random.randn(n)
income = 18000 + 3500*np.random.randn(n)
```

```python
data = pd.DataFrame({
    "Gender": gender_data,
    "Economic Status": income_data,
    "Height": height,
    "Weight": weight,
    "Age": age,
    "Income": income
})
```

```python
data.head()
```

### Agrupación de Datos

```python
grouped_gender = data.groupby("Gender") #pd.groupby sirve para dividir variables categoricas de una columna
```

```python
grouped_gender.groups #.groups permite ver los elementos que tiene cada varible categorica
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
grouped_gender.get_group("Female") # .get_group("..") nos devuelve solo los valores referidos a esa varible categorica
```

```python
doble_agrupacion = data.groupby(["Gender", "Economic Status"])
```

```python
len(doble_agrupacion)
```

```python
for names, groups in doble_agrupacion:
    print(names)
    print(groups)
```

## Operaciones sobre datos agrupados

```python
doble_agrupacion.sum() #Funcion para sumar datos agrupados
```

```python
doble_agrupacion.mean() #Funcion para obtener el promedio de datos agrupados
```

```python
doble_agrupacion.size() #Funcion para obtener el numero de elementos por cada agrupacion
```

```python
doble_agrupacion.describe() #Obtener los estadisticos basicos
```

```python
grouped_income = doble_agrupacion["Income"] #OJO CON ESTO
```

```python
grouped_income.describe()
```

```python
doble_agrupacion.aggregate({
    "Income": np.sum,
    "Age": np.mean,
    "Height": np.std
}) #pd.groupby.aggregate sirve para aplicar opraciones en datos agrupados de manera independiente para cada columna
```

```python
doble_agrupacion.aggregate({
    "Age": np.mean,
    "Height": lambda h: np.mean(h)/np.std(h)
})
```

```python
doble_agrupacion.aggregate([np.sum, np.mean, np.std]) #Se puede incluir funciones 
#Directas en cada columna del groupby con la funcion .aggregate()
```

```python
doble_agrupacion.aggregate([lambda x: np.mean(x)/np.std(x)])
```

## Transformación de Variables

```python
zscore = lambda x : ( x - x.mean())/ x.std()
doble_agrupacion.transform(zscore) #Funcion que te permite transformdar los datos de todas las columnas
# o de cualquier columna
```

```python
import matplotlib.pyplot as plt
```

```python
z_group = doble_agrupacion.transform(zscore)
```

```python
plt.hist(z_group["Age"])
```

```python
fill_na_mean = lambda x: x.fillna(x.mean()) #Funcion para llenar espacios en blanco
```

```python
doble_agrupacion.transform(fill_na_mean) #.transform() sirve para transformar
#los datos de las columnas 
```
