```python
import matplotlib.pyplot as plt
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(12, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
plt.figure(figsize=(10,5), dpi=85) #con figsize cambiamos el tamaño del grafico
#con dpi le decimos el numero de pixeles de tamaño (opcional)
plt.bar(x=names, height=values)
```

```python
figura = plt.figure(figsize=(10,5), dpi=85)#Este codigo es el contenedor de la figura
ax = figura.gca()
ax.bar(x=names, height=values)
```

```python
figura = plt.figure(figsize=(10,5), dpi=85)#Este codigo es el contenedor de la figura
ax = figura.gca() #Con esto ingreso a la clase AXES para hace uso de lo de abajo
ax.bar(x=names, height=values)
ax.set_xlabel("Grupos") #Colocando titulo al eje x
ax.set_ylabel("Cantidad") #Colocando titutlo al eje y
ax.set_title("Cantidad por Grupo") #Colocando titulo a la grafica
```
